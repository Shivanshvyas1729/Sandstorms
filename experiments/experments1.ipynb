````python
import requests
import os

# ==============================
# CONFIGURATION
# ==============================

stations = {
    "Jodhpur": "42339099999",
    "Bikaner": "42165099999",
    "Jaipur": "42348099999",
    "New_Delhi": "42182099999",
    "Ahmedabad": "42647099999"
}

start_year = 2010
end_year = 2023

base_url = "https://www.ncei.noaa.gov/data/global-hourly/access"

# 🔹 MAIN OUTPUT FOLDER
main_folder = "RAW_DATA1/NOAA_DATA1"
os.makedirs(main_folder, exist_ok=True)

# ==============================
# DOWNLOAD FUNCTION
# ==============================

for station_name, station_id in stations.items():

    print(f"\nDownloading data for {station_name}...")

    # Create station subfolder inside main folder
    station_folder = os.path.join(main_folder, station_name)
    os.makedirs(station_folder, exist_ok=True)

    for year in range(start_year, end_year + 1):

        url = f"{base_url}/{year}/{station_id}.csv"
        filename = os.path.join(
            station_folder,
            f"{station_name}_{year}.csv"
        )

        try:
            response = requests.get(url, timeout=30)

            if response.status_code == 200 and len(response.content) > 1000:
                with open(filename, "wb") as f:
                    f.write(response.content)
                print(f"✔ {year} downloaded")
            else:
                print(f"✘ {year} not available")

        except Exception as e:
            print(f"⚠ Error for {year}: {e}")

print("\nAll downloads completed ✅")
````

````python
# ==========================================
# MERGE NOAA FILES WITH CITY & STATE
# ==========================================

import pandas as pd
import os
from glob import glob

# ✅ Base folder (change if needed)
base_path = "RAW_DATA1/NOAA_DATA1"

# ✅ Station metadata (Professional way)
station_metadata = {
    "Jodhpur": {
        "City": "Jodhpur",
        "State": "Rajasthan"
    },
    "Bikaner": {
        "City": "Bikaner",
        "State": "Rajasthan"
    },
    "Jaipur": {
        "City": "Jaipur",
        "State": "Rajasthan"
    },
    "New_Delhi": {
        "City": "New Delhi",
        "State": "Delhi"
    },
    "Ahmedabad": {
        "City": "Ahmedabad",
        "State": "Gujarat"
    }
}

station_dfs = []

print("Starting merge process...\n")

for station, meta in station_metadata.items():

    folder_path = os.path.join(base_path, station)
    files = glob(os.path.join(folder_path, "*.csv"))

    print(f"Processing {station} → {len(files)} files found")

    if len(files) == 0:
        print("⚠ No files found. Check path.")
        continue

    yearly_dfs = []

    for file in files:
        try:
            df = pd.read_csv(file, low_memory=False)

            # Add metadata columns
            df["Station_Name"] = station
            df["City"] = meta["City"]
            df["State"] = meta["State"]

            yearly_dfs.append(df)

        except Exception as e:
            print(f"⚠ Error reading {file}: {e}")

    if yearly_dfs:
        station_df = pd.concat(yearly_dfs, ignore_index=True)
        station_dfs.append(station_df)

# Final Merge
if station_dfs:

    final_df = pd.concat(station_dfs, ignore_index=True)

    # Convert DATE
    final_df["DATE"] = pd.to_datetime(final_df["DATE"], errors="coerce")

    # Add time features
    final_df["Year"] = final_df["DATE"].dt.year
    final_df["Month"] = final_df["DATE"].dt.month
    final_df["Day"] = final_df["DATE"].dt.day
    final_df["Hour"] = final_df["DATE"].dt.hour

    print("\nAll stations merged successfully ✅")
    print("Final Shape:", final_df.shape)

    print("\nColumns:\n")
    print(final_df.columns)

else:
    print("❌ No data loaded.")
````

````python
import pandas as pd
import numpy as np

df = clean_df.copy()

# ======================================================
# 1️⃣ WIND SPEED EXTRACTION
# ======================================================

# wind_raw format:
# "Direction,Quality,Type,Speed,SpeedQuality"
# Example: "090,1,N,0010,1"
# Means:
# Direction = 90°
# Speed = 0010 → 1.0 m/s
# We divide by 10 because ISD stores speed × 10

def get_wind_speed(x):
    try:
        # 4th element is wind speed
        return float(x.split(",")[3]) / 10
    except:
        return np.nan

df["wind_speed"] = df["wind_raw"].apply(get_wind_speed)

# Why needed?
# Dust storms are triggered when wind speed exceeds threshold (~7–8 m/s)


# ======================================================
# 2️⃣ TEMPERATURE EXTRACTION
# ======================================================

# temperature_raw example:
# "+0120,1"
# +0120 → 12.0°C
# Divide by 10

def get_temp(x):
    try:
        return float(x.split(",")[0]) / 10
    except:
        return np.nan

df["temperature"] = df["temperature_raw"].apply(get_temp)

# Why needed?
# High surface heating → convection → instability → strong winds


# ======================================================
# 3️⃣ DEWPOINT EXTRACTION
# ======================================================

df["dewpoint"] = df["dewpoint_raw"].apply(get_temp)

# Why needed?
# Used to compute humidity.
# Low humidity favors dust lifting.


# ======================================================
# 4️⃣ VISIBILITY EXTRACTION
# ======================================================

# visibility_raw example:
# "002000,1,N,1"
# First part = 2000 meters

def get_visibility(x):
    try:
        return float(x.split(",")[0])
    except:
        return np.nan

df["visibility"] = df["visibility_raw"].apply(get_visibility)

# Why needed?
# Dust storms reduce visibility sharply.
# Very strong indicator for dust event.


# ======================================================
# 5️⃣ SEA LEVEL PRESSURE EXTRACTION
# ======================================================

# sea_level_pressure_raw example:
# "10158,1"
# Means 1015.8 hPa

def get_slp(x):
    try:
        return float(x.split(",")[0]) / 10
    except:
        return np.nan

df["pressure"] = df["sea_level_pressure_raw"].apply(get_slp)

# Why needed?
# Pressure drop → strong pressure gradient → high winds.


# ======================================================
# 6️⃣ RELATIVE HUMIDITY CALCULATION
# ======================================================

# Using Magnus formula
# Converts temperature & dewpoint into humidity %

df["relative_humidity"] = 100 * (
    np.exp((17.625 * df["dewpoint"]) / (243.04 + df["dewpoint"])) /
    np.exp((17.625 * df["temperature"]) / (243.04 + df["temperature"]))
)

# Why needed?
# Dust lifting more likely in dry air (low humidity).


# ======================================================
# 7️⃣ WEATHER CODE EXTRACTION (TARGET)
# ======================================================

# manual_weather_1 example:
# "05,1"
# First number = WMO weather code

def get_ww(x):
    try:
        return int(x.split(",")[0])
    except:
        return np.nan

df["WW_code"] = df["manual_weather_1"].apply(get_ww)

# Dust related WMO codes
dust_codes = [6,7,8,9] + list(range(30,40)) + [95,98]

df["Dust_Storm"] = df["WW_code"].apply(
    lambda x: 1 if x in dust_codes else 0
)

# Why needed?
# This becomes our supervised learning label.
````

````python
# Convert datetime to proper format
df["datetime_utc"] = pd.to_datetime(df["datetime_utc"])

# Extract date only
df["date"] = df["datetime_utc"].dt.date

# ----------------------------
# IMPROVED DAILY AGGREGATION
# ----------------------------

daily_df = df.groupby(["city", "date"]).agg({

    # Wind: strongest wind of the day (trigger variable)
    "wind_speed": "max",

    # Temperature: highest temperature (max heating)
    "temperature": "max",

    # Humidity: lowest humidity (dry condition favors dust)
    "relative_humidity": "min",

    # Pressure: lowest pressure (falling pressure system)
    "pressure": "min",

    # Visibility: lowest visibility
    "visibility": "min",

    # Dust label: if any hour dust → dust day
    "Dust_Storm": "max"

}).reset_index()

print("Improved Daily Dataset Shape:", daily_df.shape)
print(daily_df.head())
````

Hourly label:

“Did dust occur at 3 PM?”

Daily label:

“Was this a dust day?”

These are different modeling problems.

Daily modeling is:
✔ More stable
✔ More realistic for forecasting
✔ Easier to combine with ERA5
